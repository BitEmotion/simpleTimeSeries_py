```python
def generateMA2():
    x0 = 2;
    x1 = 3;
    xt = 0;

    a_t = uniform(-1,1) 
    a_t_minus_1 = uniform(-1,1) 
    a_t_minus_2 = uniform(-1,1) 

    MA2 = [x0,x1]

    for i in range(2,1440):
        x_t = a_t + (0.7 * a_t_minus_1) + (0.8 * a_t_minus_2)

        a_t_minus_2 = a_t_minus_1
        a_t_minus_1 = a_t
        a_t = uniform(-1,1) 
    
        MA2.append(x_t)
        
    return MA2
```

```python
MA2 = generateMA2()

MA2_ACF_DATA = []
for i in range(0,140):
    MA2_ACF_DATA.append(autoCorrelation(MA2, i+1))
    
print(MA2_ACF_DATA)
```
