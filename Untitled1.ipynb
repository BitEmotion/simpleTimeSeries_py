```python
import numpy as np
import matplotlib.pyplot as plt
from random import *
```

```python
# MA2 
```

```python
x0 = 2;
x1 = 3;
xt = 0;

a_t = uniform(-1,1) 
a_t_minus_1 = uniform(-1,1) 
a_t_minus_2 = uniform(-1,1) 

MA2 = [x0,x1]

for i in range(2,1440):
    x_t = a_t + (0.7 * a_t_minus_1) + (0.8 * a_t_minus_2)

    a_t_minus_2 = a_t_minus_1
    a_t_minus_1 = a_t
    a_t = uniform(-1,1) 
    
    MA2.append(x_t)
    
print(MA2)
```

```python
plt.plot(MA2)
plt.show()
```

```python
# AR2
```

```python
x_t_minus_2 = 2
x_t_minus_1 = 3
x_t = 0

a_t = 0

AR2 = [x_t_minus_2, x_t_minus_1]
for i in range(2,1440):
    a_t =  uniform(-1,1) 
    
    x_t = (1.5 * x_t_minus_1) - (0.75 * x_t_minus_2) + a_t
    x_t_minus_2 = x_t_minus_1
    x_t_minus_1 = x_t
    
    AR2.append(x_t)

print(AR2);
```

```python
plt.plot(AR2)
plt.show()
```
